# 1. Write a simple function


```python
def greet(name):
    print(f"Hello, {name}!")

# Example usage
greet('world')  # Output: Hello, world!
```

Explanation:
The function greet(name) takes a single parameter name (a string).
It uses an f-string (f"Hello, {name}!") to format and print the greeting message.
When you call greet('world'), it prints "Hello, world!".

# 2. If/else statements


```python
def goldilocks(bed_length):
    if bed_length < 140:
        print("Too small!")
    elif bed_length > 150:
        print("Too large!")
    else:
        print("Just right. :)")

# Example test cases
goldilocks(139)  # Output: Too small!
goldilocks(140)  # Output: Just right. :)
goldilocks(151)  # Output: Too large!
goldilocks(150)  # Output: Just right. :)
```

# 3. For loops


```python
def square_list(numbers):
    squared_numbers = []  # Create an empty list to store squared values
    for num in numbers:   # Loop through each number in the input list
        squared_numbers.append(num ** 2)  # Square the number and add it to the list
    return squared_numbers  # Return the new list

# Example usage
print(square_list([1, 2, 3]))  # Output: [1, 4, 9]

# extra examples runs
print(square_list([4, 5, 6]))  # Output: [16, 25, 36]
print(square_list([-1, -2, -3]))  # Output: [1, 4, 9]
print(square_list([0, 10, 100]))  # Output: [0, 100, 10000]

```

Explanation
The function takes a list of numbers as input.
It loops through each number in the list.
It squares each number (num ** 2).
It stores the squared values in a new list.
It returns the new list as output.

# 4. While loops


```python
def fibonacci_stop(max_value):
    fibonacci_sequence = [1, 1]  # Start with the first two Fibonacci numbers
    while True:
        next_number = fibonacci_sequence[-1] + fibonacci_sequence[-2]  # Sum of last two numbers
        if next_number >= max_value:  # Stop if the next number exceeds max_value
            break
        fibonacci_sequence.append(next_number)  # Add the next number to the sequence
    return fibonacci_sequence

# Example test case
print(fibonacci_stop(30))  # Output: [1, 1, 2, 3, 5, 8, 13, 21]

# extra examples runs
print(fibonacci_stop(50))  # Output: [1, 1, 2, 3, 5, 8, 13, 21, 34]
print(fibonacci_stop(10))  # Output: [1, 1, 2, 3, 5, 8]
print(fibonacci_stop(5))   # Output: [1, 1, 2, 3]
```

Explanation
Initialize the sequence with [1, 1] because the Fibonacci sequence starts with 1, 1.
Use a while loop to keep generating the next number as the sum of the last two numbers.
Stop the loop if the next number reaches or exceeds max_value.
Return the Fibonacci sequence that remains under the given maximum value.

# 5. Logical operators
  

```python
def clean_pitch(pitch_angles, status_signals):
    cleaned_list = []
    
    for angle, status in zip(pitch_angles, status_signals):
        if status > 0 and (angle < 0 or angle > 90):
            cleaned_list.append(-999)  # Set bad values to -999
        else:
            cleaned_list.append(angle)  # Keep good values unchanged
    
    return cleaned_list

# Example test case
x = [-1, 2, 6, 95]  # "Raw" pitch angles
status = [1, 0, 0, 0]  # Status signal

print(clean_pitch(x, status))  # Output: [-999, 2, 6, 95]

# extra examples runs
print(clean_pitch([0, 91, -5, 45], [0, 1, 1, 0]))  # Output: [0, -999, -999, 45]
print(clean_pitch([10, 50, 80, 100], [1, 0, 1, 1]))  # Output: [10, 50, 80, -999]
```
